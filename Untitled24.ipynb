Q2: Differentiate between one-tailed and two-tailed tests.
One-tailed test:
Used when the hypothesis is testing for the possibility of an effect in one direction only (either greater than or less than).
Example: Testing if the average score is greater than 50.
Two-tailed test:
Used when the hypothesis is testing for the possibility of an effect in both directions (either greater than or less than).
Example: Testing if the average score is different from 50 (it could be either greater or less).
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
Type 1 Error (False Positive):
Occurs when the null hypothesis is rejected when it is actually true.
Example: Concluding that a new drug is effective when it actually is not.
Type 2 Error (False Negative):
Occurs when the null hypothesis is not rejected when it is actually false.
Example: Concluding that a new drug is not effective when it actually is.
