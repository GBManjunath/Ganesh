Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask is a lightweight and flexible web framework for Python that allows developers to build web applications quickly and with minimal overhead. It is based on the WSGI (Web Server Gateway Interface) toolkit and Jinja2 template engine. Flask is designed to be simple and easy to extend, making it ideal for small to medium-sized applications.

Advantages of Flask Framework:
Lightweight and Minimalistic: Flask doesn't force any particular project structure or dependencies on developers. It's simple and provides the basics for web development, allowing developers to choose libraries or tools that best fit the project's needs.
Flexible: Flask allows you to choose your tools, libraries, and database options. It provides great flexibility for designing APIs and web applications.
Built-in Development Server: Flask comes with a built-in development server, which makes testing and debugging applications easy.
Easy to Learn: With its straightforward design and small codebase, Flask is relatively easy to learn and use, especially for beginners.
Extensible: Flask allows you to use extensions to add functionalities like authentication, form validation, and more.
Great Documentation: Flask has excellent documentation that can help developers to get started quickly.
Ideal for Prototyping: Flask is a great choice for quickly building prototypes and small applications due to its simplicity and flexibility.
Q2. Create a simple Flask application to display ‘Hello World!!’.
Here is a simple Flask application to display "Hello World!!":

python
Copy code
from flask import Flask

# Create an instance of the Flask class
app = Flask(__name__)

# Define a route for the home page
@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(debug=True)

Q3. What is App routing in Flask? Why do we use app routes?
App Routing in Flask refers to the mechanism of mapping specific URLs to functions (or views). Routes in Flask are defined using the @app.route() decorator, which tells Flask which URL should trigger a specific function.

Why do we use app routes?
URL Mapping: Routes define how the application responds to different URL patterns. They are used to link a URL to a specific view function.
Function Execution: When a user visits a URL, the corresponding route is executed to return the appropriate response, such as HTML, JSON, or other content.
Organize the Application: Routes help to logically organize the application by dividing it into different parts that respond to specific URLs.
Example:
python
Copy code
@app.route('/')
def home():
    return "This is the home page"

@app.route('/about')
def about():
    return "This is the about page"
In this example, / will trigger the home() function, and /about will trigger the about() function.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
url_for() is a function in Flask that is used for URL building. It generates the URL for a specific function, making it easy to link to dynamic URLs in templates or view functions. It is often used to avoid hardcoding URLs and makes the code more maintainable.

Usage of url_for():
The url_for() function dynamically generates the URL for the given view function, which allows us to create URLs without explicitly writing them.
The function can be used for redirecting to a specific route or linking to a route in HTML templates.
Example Code:
python
Copy code
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/redirect_to_about')
def redirect_to_about():
    return f'Redirecting to About page: {url_for("about")}'

if __name__ == '__main__':
    app.run(debug=True)
Explanation:
url_for("about") generates the URL for the about view function, which is /about.
When you visit the /redirect_to_about URL, it will display the URL of the "About" page: /about.
You can also use url_for() in HTML templates to link to other routes.

In HTML Template:
html
Copy code
<a href="{{ url_for('about') }}">About Us</a>
This will generate the correct URL for the "About" page, even if the URL structure changes later.
