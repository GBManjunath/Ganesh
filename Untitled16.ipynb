Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Seaborn is a powerful Python library for data visualization built on top of Matplotlib. It provides several plot types that are easier to use and more aesthetically pleasing than the standard Matplotlib plots. Here are five commonly used plots in Seaborn and their uses:

Line Plot (sns.lineplot)

Use: Line plots are used to visualize data points in a time series or any other ordered data. It is used to identify trends and patterns over a continuous domain (e.g., time).
Box Plot (sns.boxplot)

Use: Box plots display the distribution of a dataset through quartiles. It highlights the median, upper, and lower quartiles, as well as potential outliers. It's useful for comparing distributions across categories.
Histogram (sns.histplot)

Use: Histograms are used to display the distribution of a single numerical variable. They are useful for understanding the frequency distribution of a continuous variable.
Heatmap (sns.heatmap)

Use: Heatmaps are used to visualize the matrix of data where the individual values are represented by the color of the grid cells. This is helpful in understanding correlations or relationships between different variables.
Pair Plot (sns.pairplot)

Use: Pair plots are used to visualize pairwise relationships between multiple variables in a dataset. It’s useful for understanding how variables correlate with each other and for detecting patterns.
Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
To plot the line plot for the given dataset, you can use Seaborn's sns.lineplot function. Here’s how you can load the fmri dataset and create the plot:

python
Copy code
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title("Line Plot of Signal vs Timepoint for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
In this code:

hue="event" colors the lines based on the different events.
style="region" uses different line styles for different regions.

Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
You can use Seaborn's sns.histplot to plot the histogram of the price column, with color differentiation by the cut column:

python
Copy code
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram for 'price' with 'cut' as hue
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack", kde=True)

# Add title and labels
plt.title("Histogram of Price by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")

# Show the plot
plt.show()
In this code:

hue="cut" adds color differentiation based on the cut of the diamonds.
multiple="stack" stacks the bars for each category of cut to better visualize the distribution.
kde=True adds a kernel density estimate curve on top of the histogram.

Q6: Use the "flights" dataset from seaborn to plot a heatmap.
To create a heatmap, we first need to load the flights dataset and then use Seaborn's sns.heatmap function. Here's how you can do it:

python
Copy code
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the data to create a heatmap-friendly format
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Plot the heatmap
sns.heatmap(flights_pivot, annot=True, cmap="YlGnBu", fmt="d")

# Add title and labels
plt.title("Heatmap of Flight Passengers Over Time")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
In this code:

We use pivot_table() to reshape the data, making it easier to visualize as a heatmap.
annot=True displays the values in each cell of the heatmap.
cmap="YlGnBu" specifies the color map.
fmt="d" ensures the values are displayed as integers.
