Q1: Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
To plot a scatter plot using Plotly Express for the age and fare columns from the Titanic dataset, you can follow the steps below:

python
Copy code
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a scatter plot using Plotly Express
fig = px.scatter(titanic, x='age', y='fare', title="Scatter Plot of Age vs Fare", labels={'age': 'Age', 'fare': 'Fare'})

# Show the plot
fig.show()
In this code:

We load the titanic dataset using Seaborn.
We then use Plotly Express' scatter() function to create a scatter plot with the age column on the x-axis and the fare column on the y-axis.

Q3: Using the tips dataset in the Plotly library, plot a histogram for x = "sex" and y = "total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
To create the desired histogram with sex on the x-axis and total_bill on the y-axis, and using smoker for pattern_shape and day for color, you can use the following code:

python
Copy code
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create the histogram
fig = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker",
                   title="Histogram of Total Bill by Sex", labels={"sex": "Sex", "total_bill": "Total Bill"})

# Show the plot
fig.show()
In this code:

The x="sex" and y="total_bill" define the columns to be used in the histogram.
The color="day" differentiates the bars based on the day of the week.
The pattern_shape="smoker" differentiates the bars based on whether the person is a smoker or not.
Q4: Using the iris dataset in the Plotly library, plot a scatter matrix plot, using the "species" column for the color parameter.
To create a scatter matrix plot (also called pair plot) for the iris dataset with species as the color parameter, we can use Plotly Express' scatter_matrix() function:

python
Copy code
import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Create the scatter matrix plot
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                        color="species", title="Scatter Matrix of Iris Dataset")

# Show the plot
fig.show()
In this code:

The dimensions parameter is used to specify the columns for the scatter matrix plot.
The color="species" parameter colors the points based on the species of the iris flower.
Q5: What is Distplot? Using Plotly express, plot a distplot.
In Seaborn, distplot is used to plot univariate distributions. However, Plotly Express doesn't have a direct distplot function. Instead, you can use px.histogram() with a density plot, which provides similar functionality.

Hereâ€™s an example of how you can create a distribution plot using Plotly Express:

python
Copy code
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a histogram with a density curve (similar to a distplot)
fig = px.histogram(tips, x="total_bill", nbins=30, marginal="density", title="Distribution of Total Bill")

# Show the plot
fig.show()
In this code:

The marginal="density" option adds a kernel density estimate (KDE) on top of the histogram, which is similar to the density curve in a Seaborn distplot.
The nbins=30 argument specifies the number of bins in the histogram.
