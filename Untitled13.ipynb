Q2. Re-index the DataFrame with a new index starting from 1 and incrementing by 2
Here's the Python function to re-index the DataFrame df with a new index starting from 1 and incrementing by 2:

python
Copy code
import pandas as pd

def reindex_dataframe(df):
    df.index = range(1, 2 * len(df), 2)
    return df

# Example

df = reindex_dataframe(df)
print(df)

Q10. Create a 'MovingAverage' column for the past 7 days' sales
Here’s the Python function to calculate the moving average of sales for the past 7 days:

python
Copy code
import pandas as pd

def calculate_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7).mean()
    return df

# Example
df = pd.DataFrame({
    'Sales': [100, 200, 300, 400, 500, 600, 700, 800, 900],
    'Date': pd.date_range('2023-01-01', periods=9, freq='D')
})
df = calculate_moving_average(df)
print(df)

Q12. Select rows between specific dates
Here’s the Python function to select rows between '2023-01-01' and '2023-01-31':

python
Copy code
import pandas as pd

def select_date_range(df):
    df['Date'] = pd.to_datetime(df['Date'])
    return df[(df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')]

# Example
df = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-10', '2023-02-01', '2023-01-20'],
    'Values': [100, 200, 300, 400]
})
filtered_df = select_date_range(df)
print(filtered_df)
