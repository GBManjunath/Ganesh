Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
A scatter plot is a type of data visualization where each data point is represented as a dot on a 2D plane. The position of each dot corresponds to the values of two variables (one on the x-axis and one on the y-axis). It is used to visually represent the correlation or relationship between the two variables.

Here is how you can generate a scatter plot with the given data and add a title, x-label, and y-label:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

# Generating random data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Scatter plot
plt.scatter(x, y)

# Adding title and labels
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
This will create a scatter plot with the generated x and y values. The plot will also include a title and labels for both the x and y axes.

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to define the layout of subplots in a grid. The function takes three parameters: the number of rows, the number of columns, and the index of the current subplot (starting from 1).

The purpose of using subplot() is to show multiple plots within the same figure for comparison.

Here is the code to create four line plots using subplot():

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, first subplot
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, second subplot
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, third subplot
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, fourth subplot
plt.plot(x, y4)
plt.title("Line 4")

# Show the plot
plt.tight_layout()  # Adjust spacing between plots
plt.show()
This will display four line plots arranged in a 2x2 grid. Each subplot will have its own title.

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
A bar plot is used to represent categorical data with rectangular bars where the length of each bar is proportional to the value it represents. It is commonly used to compare different categories.

Here’s how to plot both a vertical bar plot and a horizontal bar plot using the provided data:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit)
plt.title("Profit by Company (Vertical)")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit)
plt.title("Profit by Company (Horizontal)")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
The vertical bar plot uses plt.bar().
The horizontal bar plot uses plt.barh().
Both plots compare the profits of different companies.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
A box plot (also called a box-and-whisker plot) is used to visualize the distribution of data, including the median, quartiles, and outliers. The box represents the interquartile range (IQR), with the median as a line inside the box. The "whiskers" extend from the box to show the range of the data, and any outliers are plotted individually.

Here’s how to create a box plot using the provided data:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.boxplot([box1, box2], labels=["Box1", "Box2"])

# Adding title and labels
plt.title("Box Plot Example")
plt.xlabel("Box")
plt.ylabel("Value")

# Show the plot
plt.show()
The box plot is useful to compare the distributions of two sets of data (box1 and box2 in this case), and it provides insights into the median, spread, and any potential outliers in the data.
