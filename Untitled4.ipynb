Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode of file opening.
The open() function is used to open a file in Python. It returns a file object, which can be used to read from or write to the file.

Different modes of opening a file:
'r' (Read mode): Opens the file for reading. The file pointer is placed at the beginning of the file. If the file does not exist, it raises a FileNotFoundError.

'w' (Write mode): Opens the file for writing. If the file exists, it truncates the file to zero length (erases existing content). If the file does not exist, it creates a new file.

'a' (Append mode): Opens the file for appending. Data will be written to the end of the file, and if the file doesn't exist, it will create a new one.

'x' (Exclusive creation mode): Opens the file for exclusive creation. If the file already exists, it raises a FileExistsError. If the file does not exist, it creates a new file.

'b' (Binary mode): Opens the file in binary mode. This is typically used for non-text files like images, videos, etc. It can be combined with other modes like 'rb' or 'wb'.

't' (Text mode): This is the default mode. It opens the file in text mode, meaning the file is treated as containing text, and newline characters are handled automatically. It can be combined with other modes like 'rt' or 'wt'.

'r+' (Read and write mode): Opens the file for both reading and writing. The file pointer is placed at the beginning of the file. If the file does not exist, it raises a FileNotFoundError.

'w+' (Write and read mode): Opens the file for both reading and writing. It truncates the file to zero length if it exists. If the file does not exist, it creates a new file.

'a+' (Append and read mode): Opens the file for both reading and appending. The data will be written to the end of the file, and if the file does not exist, it creates a new one.

Q2. Why close() function is used? Why is it important to close a file?
The close() function is used to close an open file. It ensures that any changes made to the file are saved, and the file resources are freed up.

Importance of closing a file:
Resource Management: When a file is opened, the operating system allocates resources like memory. Closing the file releases these resources.
Data Integrity: If a file is open and not closed, any changes made (e.g., writing data) might not be properly saved, leading to data corruption or loss.
Prevents Memory Leaks: Leaving files open consumes memory and system resources. Closing them properly helps avoid memory leaks.
Q3. Write a Python program to create a text file, write ‘I want to become a Data Scientist’ in that file, and then close the file. Open this file and read the content.
python
Copy code
# Creating and writing to a text file
file = open("data_scientist.txt", "w")  # Open file in write mode
file.write("I want to become a Data Scientist")
file.close()  # Close the file after writing

# Reading the content of the file
file = open("data_scientist.txt", "r")  # Open file in read mode
content = file.read()  # Read the content
print(content)  # Display the content of the file
file.close()  # Close the file after reading
Explanation:

The file is opened in "w" (write) mode to create the file and write content to it.
After writing, we close the file using close() to save the content.
Then, the file is opened in "r" (read) mode to read the content and display it.
Q4. Explain the following with Python code: read(), readline(), and readlines().
read(): Reads the entire content of the file as a string.
readline(): Reads a single line from the file.
readlines(): Reads all lines from the file and returns them as a list of strings.
Example:
python
Copy code
# Sample text file creation for demonstration
file = open("sample.txt", "w")
file.write("Hello, World!\nWelcome to Python.\nLet's learn file operations.")
file.close()

# Using read() to read the entire file
file = open("sample.txt", "r")
content = file.read()  # Reads the entire content
print("read():", content)
file.close()

# Using readline() to read one line at a time
file = open("sample.txt", "r")
line1 = file.readline()  # Reads the first line
line2 = file.readline()  # Reads the second line
print("readline():", line1.strip(), line2.strip())
file.close()

# Using readlines() to read all lines and store them in a list
file = open("sample.txt", "r")
lines = file.readlines()  # Reads all lines into a list
print("readlines():", [line.strip() for line in lines])  # Stripping newlines
file.close()
Explanation:

read() reads all content from the file at once.
readline() reads one line at a time.
readlines() reads all lines and stores them in a list.
Q5. Explain why the with statement is used with open(). What is the advantage of using with statement and open() together?
The with statement simplifies file handling by automatically managing resources. It ensures that the file is closed when the block of code under the with statement is completed, even if an exception occurs.

Advantages:
Automatic File Closure: No need to explicitly call close(), as with handles it automatically.
Cleaner Code: Reduces the chances of leaving files open inadvertently.
Exception Handling: If an error occurs inside the with block, it will still ensure the file is closed properly.
Example:
python
Copy code
# Using 'with' statement to open and read a file
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)  # File is automatically closed after this block
Explanation:

The with statement ensures that the file is closed automatically when the block is exited, making the code more reliable and clean.
Q6. Explain the write() and writelines() functions. Give a suitable example.
write(): Writes a string to the file. It does not add a newline after writing unless explicitly specified.
writelines(): Writes a sequence of strings (usually a list of strings) to the file. It does not add newlines unless the strings in the sequence contain them.
Example:
python
Copy code
# Using write()
with open("write_example.txt", "w") as file:
    file.write("Hello, this is a test write.\n")
    file.write("Let's learn file operations in Python.")

# Using writelines()
lines = ["Hello, this is a test writelines.\n", "Let's learn file operations in Python.\n"]
with open("writelines_example.txt", "w") as file:
    file.writelines(lines)

# Reading the content to verify
with open("write_example.txt", "r") as file:
    print(file.read())

with open("writelines_example.txt", "r") as file:
    print(file.read())
Explanation:

write() writes a single string, and you need to manually include newline characters if needed.
writelines() writes multiple strings at once, where each string in the list should already contain the desired line breaks if needed.
