Q1. What is multiprocessing in Python? Why is it useful?
Multiprocessing in Python allows a program to run multiple processes concurrently. Each process has its own Python interpreter and memory space, so it can run independently and simultaneously with other processes, making it useful for CPU-bound tasks.

Why is it useful?

Parallelism: It allows parallel execution on multiple CPUs or cores. This is particularly beneficial for CPU-bound tasks, where you need to perform heavy computations, as each process can run on a separate CPU core.
Bypass GIL: In Python, the Global Interpreter Lock (GIL) limits multi-threading for CPU-bound tasks. However, with multiprocessing, each process runs independently and has its own memory space, so the GIL does not interfere.
Q2. What are the differences between multiprocessing and multithreading?
Aspect	Multiprocessing	Multithreading
Definition	Running multiple processes concurrently.	Running multiple threads concurrently within a single process.
Concurrency	True parallelism, each process runs in its own memory space.	Concurrency but limited by GIL (only one thread executes at a time in CPython for CPU-bound tasks).
Use Case	Best for CPU-bound tasks (heavy computations).	Best for I/O-bound tasks (e.g., network requests, file I/O).
Memory Space	Each process has its own memory space.	All threads share the same memory space.
Overhead	Higher overhead as processes are heavyweight.	Lower overhead as threads are lightweight.
Python's GIL	Not affected by GIL, can run on multiple CPU cores.	Affected by the GIL, limiting the execution of threads on multiple cores for CPU-bound tasks.
Q3. Write a Python code to create a process using the multiprocessing module.
python
Copy code
import multiprocessing

# Define the target function for the process
def print_square(number):
    print(f"Square of {number}: {number ** 2}")

# Create a new process
if __name__ == "__main__":
    process = multiprocessing.Process(target=print_square, args=(5,))  # Pass the argument 5
    process.start()  # Start the process
    process.join()   # Wait for the process to finish
Explanation:

The multiprocessing.Process class is used to create a new process.
The target argument specifies the function to run in the new process.
The args argument is used to pass arguments to the function.
The start() method starts the process, and join() waits for the process to finish.
Q4. What is a multiprocessing pool in Python? Why is it used?
A multiprocessing pool in Python is a collection of worker processes that can execute tasks concurrently. It is useful for managing a pool of workers to perform tasks in parallel without manually managing the individual processes. This is especially helpful when you need to run the same function multiple times across different input data.

Why is it used?

Efficiency: It simplifies parallel processing by managing a pool of worker processes.
Scalability: It automatically handles load balancing and the number of worker processes needed.
Performance: It speeds up CPU-bound tasks by leveraging multiple CPU cores.
Q5. How can we create a pool of worker processes in Python using the multiprocessing module?
To create a pool of worker processes in Python, you can use the multiprocessing.Pool class. Here's how you can do it:

python
Copy code
import multiprocessing

# Define the function that will be executed by each worker
def square(number):
    return number ** 2

# Create a pool of worker processes
if __name__ == "__main__":
    pool = multiprocessing.Pool(processes=4)  # Create a pool with 4 processes

    # Use map() to distribute the workload across the pool
    numbers = [1, 2, 3, 4, 5]
    result = pool.map(square, numbers)  # Distributes the function across the numbers
    print(result)  # Output: [1, 4, 9, 16, 25]

    pool.close()  # Close the pool
    pool.join()   # Wait for all worker processes to finish
Explanation:

The multiprocessing.Pool(processes=N) creates a pool with N worker processes.
The map() function applies the square() function to each element in the list numbers across the available pool of workers.
The close() method is used to prevent any more tasks from being added to the pool.
The join() method ensures the program waits for all worker processes to complete before continuing.
Q6. Write a Python program to create 4 processes, each process should print a different number using the multiprocessing module in Python.
python
Copy code
import multiprocessing

# Define the target function for each process
def print_number(number):
    print(f"Number: {number}")

if __name__ == "__main__":
    # Create 4 processes, each printing a different number
    process1 = multiprocessing.Process(target=print_number, args=(1,))
    process2 = multiprocessing.Process(target=print_number, args=(2,))
    process3 = multiprocessing.Process(target=print_number, args=(3,))
    process4 = multiprocessing.Process(target=print_number, args=(4,))

    # Start the processes
    process1.start()
    process2.start()
    process3.start()
    process4.start()

    # Join the processes
    process1.join()
    process2.join()
    process3.join()
    process4.join()
Explanation:

This program creates 4 processes, each one running the print_number function with a different argument.
Each process is started using start() and the main program waits for their completion using join().
