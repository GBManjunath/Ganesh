Q1. What is the Probability Density Function?
The Probability Density Function (PDF) describes the likelihood of a continuous random variable taking a particular value. It is the continuous counterpart to the probability mass function (PMF) used for discrete distributions. The PDF is a function that, when integrated over an interval, gives the probability that a random variable falls within that interval. For a continuous distribution, the probability of the variable taking any exact value is technically zero, but the area under the curve between two values represents the probability.

Q2. What are the types of Probability Distribution?
There are two main types of probability distributions:

Discrete Probability Distributions: These describe distributions of discrete random variables, where the outcomes are countable. Examples include:

Binomial Distribution: Represents the number of successes in a fixed number of independent trials with the same probability of success.
Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given the average rate.
Continuous Probability Distributions: These describe distributions of continuous random variables, where the outcomes are uncountably infinite. Examples include:

Normal Distribution: A bell-shaped curve representing the distribution of a continuous random variable with a known mean and variance.
Exponential Distribution: Describes the time between events in a Poisson process, where events occur continuously and independently at a constant average rate.
Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.
Here's a Python function to calculate the PDF of a normal distribution:

python
Copy code
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution.
    
    Parameters:
    x (float): The point at which to calculate the PDF.
    mean (float): The mean (mu) of the distribution.
    std_dev (float): The standard deviation (sigma) of the distribution.
    
    Returns:
    float: The value of the PDF at point x.
    """
    exponent = - (x - mean) ** 2 / (2 * std_dev ** 2)
    return (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(exponent)

# Example usage:
print(normal_pdf(0, 0, 1))  # Standard normal distribution at x = 0
Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.
Properties of Binomial Distribution:

Fixed number of trials: The experiment is repeated a fixed number of times (n).
Two possible outcomes: Each trial results in one of two outcomes (success or failure).
Constant probability: The probability of success (p) remains constant for each trial.
Independence: Each trial is independent of the others.
Examples of events where Binomial distribution can be applied:

Flipping a coin: The number of heads in a series of 10 coin flips.
Quality control: The number of defective items found in a sample of 50 manufactured items.

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with a given mean at a given point.
Here's a Python function to calculate the CDF of a Poisson distribution:

python
Copy code
from scipy.stats import poisson

def poisson_cdf(k, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.
    
    Parameters:
    k (int): The point at which to calculate the CDF.
    mean (float): The mean (lambda) of the distribution.
    
    Returns:
    float: The value of the CDF at point k.
    """
    return poisson.cdf(k, mean)

# Example usage:
print(poisson_cdf(3, 5))  # Poisson CDF for k=3, mean=5
Q7. How Binomial distribution differs from Poisson distribution?
Binomial Distribution: Used for modeling the number of successes in a fixed number of trials, each with the same probability of success. It is discrete and defined by two parameters: the number of trials (n) and the probability of success (p).

Poisson Distribution: Used for modeling the number of events occurring within a fixed interval of time or space, given the average rate of occurrence. It is also discrete but is parameterized by the rate of occurrence (λ), rather than the number of trials.

Key Differences:

Binomial is used for a fixed number of trials, while Poisson is used for a fixed interval.
Poisson can be used as an approximation to the Binomial distribution when the number of trials is large and the probability of success is small.

Q10. In normal distribution with respect to the mean position, where does the least frequent data appear?
In a normal distribution, the least frequent data appears at the extremes (far away from the mean). As you move farther from the mean, the frequency of data points decreases, with the least frequent data points being in the tails of the distribution. The peak of the normal distribution occurs at the mean, where the most frequent values are located.
