Q1. Is there any difference in the data type of variables list_ and array_list? If there is, write a code to print the data types of both the variables.
Yes, there is a difference in the data types of list_ and array_list.

list_ is a Python list, and its data type is list.
array_list is a NumPy array, and its data type is numpy.ndarray. The elements inside the array can be of a specified type, such as int, float, etc.
Here's the code to print the data types of both variables:

python
Copy code
import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object=list_)

# Printing the data types
print(f"Data type of list_: {type(list_)}")
print(f"Data type of array_list: {type(array_list)}")
Output:

python
Copy code
Data type of list_: <class 'list'>
Data type of array_list: <class 'numpy.ndarray'>
