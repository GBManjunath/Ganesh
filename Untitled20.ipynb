Q1. What are the three measures of central tendency?
The three measures of central tendency are:

Mean: The average of all the data points.
Median: The middle value when the data points are arranged in ascending or descending order.
Mode: The value that appears most frequently in the dataset.
Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?
Mean: The arithmetic average of a dataset, calculated by summing all values and dividing by the number of values. The mean is sensitive to extreme values (outliers).
Median: The middle value of a dataset when the data is arranged in order. It divides the data into two equal halves and is less influenced by outliers than the mean.
Mode: The most frequent value in the dataset. A dataset can have more than one mode or no mode at all if all values are unique.
These measures describe the "central" point around which the data tends to cluster. The mean is ideal for symmetric distributions, the median is preferred for skewed distributions, and the mode is used when identifying the most common value is important (e.g., for categorical data).

Q6. What is a Venn diagram?
A Venn diagram is a graphical representation of sets and their relationships. It uses circles to represent sets and their intersections to show common elements. It helps in visualizing operations like unions, intersections, and differences between sets.

Q8. What do you understand about skewness in data?
Skewness refers to the asymmetry of the distribution of data. A dataset can be:

Right-skewed (positively skewed): The right tail (higher values) is longer than the left tail, meaning most values are concentrated on the left.
Left-skewed (negatively skewed): The left tail (lower values) is longer than the right tail.
Symmetric: If the data is evenly distributed on both sides of the mean.
Q9. If a data is right-skewed, then what will be the position of the median with respect to the mean?
In a right-skewed (positively skewed) distribution, the mean is typically greater than the median. This is because the mean is pulled to the right by the higher values in the tail.

Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?
Covariance: Measures the degree to which two variables change together. If the covariance is positive, the variables tend to increase together; if negative, one increases as the other decreases. The value is in the units of the variables.

Correlation: A standardized version of covariance that provides a measure of the strength and direction of the relationship between two variables. It ranges from -1 to 1, where:

1 means a perfect positive linear relationship
-1 means a perfect negative linear relationship
0 means no linear relationship.

Q12. For a normal distribution data, what is the relationship between its measure of central tendency?
In a normal distribution, the mean, median, and mode are all equal and lie at the center of the distribution. The data is symmetrically distributed around the central point.

Q13. How is covariance different from correlation?
Covariance measures the direction of the linear relationship between two variables.
Correlation standardizes this measure to indicate both the strength and direction of the relationship, and is not affected by the units of measurement.
Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.
Mean: Outliers can significantly affect the mean, pulling it towards the extreme values. For example, in the dataset [1, 2, 3, 100], the mean would be much higher than the typical values due to the outlier (100).
Median and Mode: The median and mode are less affected by outliers. The median remains the same in the example above, and the mode is unaffected as it doesn't rely on extremes.
Dispersion: Outliers can increase the variance and standard deviation, making the spread appear larger than it truly is.
